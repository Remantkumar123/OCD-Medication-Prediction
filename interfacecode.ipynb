```python
#Check the null values
df.isnull().sum()
```

```python
df['Previous Diagnoses'].fillna('none',inplace=True)
```

```python
df['Medications'].fillna('none',inplace=True)
```

```python
#Check the null values
df.isnull().sum()
```

```python
#Drop the Unnecessary Columns
df.drop(columns = ['Patient ID','Ethnicity','Marital Status','Education Level','OCD Diagnosis Date','Family History of OCD','Obsession Type','Compulsion Type'],inplace = True)
df
```

```python
#List of unique values from gender
df['Gender'].unique()
```

```python
df['Gender']=df['Gender'].map({'Female':0, 'Male':1})
df
```

```python
#List of unique values from Previous Diagnoses
df['Previous Diagnoses'].unique()
```

```python
df['Previous Diagnoses']=df['Previous Diagnoses'].map({'MDD':0, 'none':1,'PTSD':2,'GAD':3,'Panic Disorder':4})
df
```

```python
#List of unique values from Depression Diagnosis
df['Depression Diagnosis'].unique()
```

```python
df['Depression Diagnosis']=df['Depression Diagnosis'].map({'Yes':0, 'No':1})
df
```

```python
#List of unique values from Anxiety Diagnosis	 
df['Anxiety Diagnosis'].unique()
```

```python
df['Anxiety Diagnosis']=df['Anxiety Diagnosis'].map({'Yes':0, 'No':1})
df
```

```python
#List of unique values from Medications		 
df['Medications'].unique()
```

```python
df['Medications']=df['Medications'].map({'SNRI':0, 'SSRI':1,  'Benzodiazepine':2, 'none':3})
df
```

```python
# Feature selection
# Replace 'features' and 'target' with the actual column names
x =df.iloc[:,0:8]
y=df.iloc[:,-1]
x
```

```python
y
```

```python
#import train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=77)
x_train
```

```python
y_train
```

```python
#Transforming the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #removes the mean and scales each feature/variable to unit variance.
x_train = scaler.fit_transform(x_train) 
x_test = scaler.transform(x_test)
x_test
```

```python
#Importing the GaussianNB algorithm
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
#fitting i the model
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
print(pred[:150])
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
# construction of heatmap for confusion matrix
mat = confusion_matrix(pred,y_test)
names = np.unique(pred)
sns.heatmap(mat,square=True,annot=True,fmt='d',cbar = False)
plt.xlabel('Truth')
plt.ylabel('Predicted')
plt.plot()
```

```python
#Calculating the accuracy
from sklearn.metrics import accuracy_score,classification_report
score_gnb = accuracy_score(pred,y_test)
print("Accuracy:",score_gnb*100,"%")
```

```python
# Save the model
import pickle
with open('gaussian_naive_bayes_model.pkl', 'wb') as f:
    pickle.dump(model, f)
print("Model saved as 'gaussian_naive_bayes_model.pkl'")
```
